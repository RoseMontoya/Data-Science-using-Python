# Transformation HandsOn 
#### An activity to practice transforming data to be normally distributed. 

## Loading in packages and data

```python
import pandas as pd
import numpy as np
```

```python
ParksnRec = pd.read_csv('Seattle_ParksnRec.csv')
ParksnRec.head()
```

## # of trips Winter
### Making a Histogram

```python
ParksnRec['# of trips Winter'].hist()
```

#### Looks a positively skewed.
### Taking the square root

```python
ParksnRec['# of trips WinterSQRT'] = np.sqrt(ParksnRec['# of trips Winter'])
ParksnRec.head()
```

```python
ParksnRec['# of trips WinterSQRT'].hist()
```

#### Gonna take a look at the variable log.

```python
ParksnRec['# of trips WinterLOG'] = np.log(ParksnRec['# of trips Winter'])
ParksnRec.head()
```

```python
ParksnRec['# of trips WinterLOG'].hist()
```

#### Square root is definitely the better option. 
## # of participants Winter
### Making a Histogram

```python
ParksnRec['# of participants Winter'].hist()
```

#### Looks to be posititvely skewed. 
### Taking the Square Root

```python
ParksnRec['# of participants WinterSQRT'] = np.sqrt(ParksnRec['# of participants Winter'])
ParksnRec.head()
```

```python
ParksnRec['# of participants WinterSQRT'].hist()
```

#### Looking good!
