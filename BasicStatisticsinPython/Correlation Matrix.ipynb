# Correlation Matrix
#### This is an activity to practice using correlation matrices in Python. 

## Import Data and Packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
powerlifting = pd.read_csv('power_lifting.csv')
powerlifting.head()
```

```python
powerlifting.info()
```

## Data Wrangling

```python
power = powerlifting.drop(['Name', 'Sex', 'Equipment','Division', 'WeightClassKg', 'Place'], axis = 1)
```

```python
power.head()
```

## Run Correlation Test

```python
power.corr(method='pearson').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('coolwarm'), axis=1)
```

### The strongest correlation is between BestSquatKg and Total Kg. This makes sense since the most you could lift is also probably your best. There is also a strong correlation between BestSquatKg and Best Bench Kg, Best Dead lift kg,  and Wilks. Then between Best Bench kg and Best Dead lift kg, best dead lift kg and total kg, and total kg and wilks. It seems the best kg are a good predictor of the other best kg, so your best dead lift will be a good predictor of your best squat and so forth. But the 4Kg are not good predictors. Wilks is the best predictor of of total Kg, so wilks is probably based off of a person total kg. Best Squat kg are the best over all predictor. Age has no real influcence, and bodyweight only has a moderate to low influence on correlation.
